```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# Loading the data
df = pd.read_csv('data_2018_2023_chl_sst_sswh.csv')
```

# Convert temperature values from Kelvin to Celsius

```python
df["analysed_sst"] = df["analysed_sst"] - 273.15
df["analysis_error"] = df["analysis_error"] - 273.15
```

# Data Types

```python
# Checking the features' data types
df.dtypes
```

# Missing Values

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Checking for any categorical features that might need encoding
categorical_features = df.select_dtypes(include=['object']).columns
print("Categorical Features for Potential Encoding:", list(categorical_features))
```

# Data Cleaning

```python
# Checking for duplicates
df[df.duplicated(keep='first')]
```

```python
# Converting 'time' to datetime
df['time'] = pd.to_datetime(df['time'], errors='coerce')
```

```python
df.dtypes
```

```python
# Dropping irrelevant columns
df.drop(columns=['time.1'], inplace=True)
df.drop(columns=['time.2'], inplace=True)
```

```python
# Converting the clean dataframe to a new CSV
df.to_csv('data_2018_2023_chl_sst_sswh_clean.csv', index=False)
```
